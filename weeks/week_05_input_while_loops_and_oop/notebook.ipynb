## Validation pattern

```python
responses = iter(["maybe", "yes"])

def fake_input(prompt: str) -> str:
    value = next(responses)
    print(f"{prompt}{value}")
    return value

answer = fake_input("Continue? (yes/no): " ).strip().lower()
while answer not in {"yes", "no"}:
    answer = fake_input("Please type yes or no: " ).strip().lower()

print("Accepted:", answer)
```

## A simple class

```python
class Person:
    def __init__(self, first_name: str, last_name: str, age: int) -> None:
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

    def label(self) -> str:
        return f"{self.first_name} {self.last_name} ({self.age})"

people = [Person("Alex", "Murphy", 21), Person("Beth", "Stone", 22)]
print([person.label() for person in people])
```
