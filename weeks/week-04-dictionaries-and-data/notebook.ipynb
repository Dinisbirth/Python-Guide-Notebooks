## Records with dictionaries

```python
people = [
    {"first_name": "Alex", "last_name": "Murphy", "age": 21},
    {"first_name": "Beth", "last_name": "Stone", "age": 22},
    {"first_name": "Alex", "last_name": "Murphy", "age": 21},
]

for person in people:
    print(person)
```

## Sets for comparison and deduplication

```python
def common_items(iterable_a, iterable_b):
    return sorted(set(iterable_a) & set(iterable_b))

unique_people = [dict(record) for record in {tuple(sorted(person.items())) for person in people}]

print(common_items([1, 2, 3, 4], [3, 4, 5, 6]))
print(unique_people)
```
