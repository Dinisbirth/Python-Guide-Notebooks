## A simple function

```python
def square(number: int) -> int:
    return number * number

print(square(5))
```

## Reusable multiples solver

```python
def sum_of_multiples(limit: int, divisors: tuple[int, ...] = (3, 5)) -> int:
    total = 0
    for value in range(limit):
        if any(value % divisor == 0 for divisor in divisors):
            total += value
    return total

print(sum_of_multiples(10))
print(sum_of_multiples(1000))
```
