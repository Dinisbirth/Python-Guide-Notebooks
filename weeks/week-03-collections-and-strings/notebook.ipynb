## List operations

```python
names = ["Alex", "Beth", "Chris"]
names.remove("Beth")
names.extend(["Dinis", "Emma"])

print(names)
print(names[:2])
print(list(reversed(names)))
```

## String manipulation

```python
sentence = "The quick brown fox jumps over the lazy dog"
words = sentence.split()

print(sentence[::-1])
print(words[1:3])
print(["".join(reversed(word)) for word in words])
```
