## Core built-ins

```python
numbers = [4, -2, 9, 1]
letters = ["a", "b", "c", "d"]
scores = [85, 90, 78, 92]

print(abs(-12))
print(all(score >= 70 for score in scores))
print(any(number < 0 for number in numbers))
print(len(numbers), max(numbers), min(numbers))
print(sorted(numbers))
print(list(zip(letters, scores)))
```

## Standard modules

```python
import math
import statistics

values = [4, 7, 9, 10, 13]
print(math.sqrt(81))
print(statistics.mean(values))
print(statistics.median(values))
```
