## Project setup

```python
from pathlib import Path
import sys

def find_project_root(start: Path | None = None) -> Path:
    current = (start or Path.cwd()).resolve()
    for candidate in [current, *current.parents]:
        if (candidate / "requirements.txt").exists() and (candidate / "weeks").exists():
            return candidate
    return current

ROOT = find_project_root()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

ROOT
```

## Correlation and engineering

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

housing = pd.read_csv(ROOT / "data" / "housing" / "housing.csv")
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

correlations = train_set.corr(numeric_only=True)["median_house_value"].sort_values(ascending=False)
display(correlations)

engineered = train_set.copy()
engineered["rooms_per_household"] = engineered["total_rooms"] / engineered["households"]
engineered["bedrooms_per_room"] = engineered["total_bedrooms"] / engineered["total_rooms"]
engineered["population_per_household"] = engineered["population"] / engineered["households"]
display(engineered.corr(numeric_only=True)["median_house_value"].sort_values(ascending=False).head(10))

imputer = SimpleImputer(strategy="median")
numeric_housing = train_set.drop(columns=["ocean_proximity"])
imputed = imputer.fit_transform(numeric_housing)
print(imputed.shape)
```
