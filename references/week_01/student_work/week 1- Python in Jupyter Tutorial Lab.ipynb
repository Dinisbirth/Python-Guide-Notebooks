# **Section 4: Functions**

**Write a func on max_of_two that takes two numbers as arguments and returns the larger of 
them.**

```python
def max_of_two(a, b):
    if a > b:
        return a
    else:
        return b

print("Max of 10 and 20:", max_of_two(10, 20))
```

```python
def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total


print("Max of 10 and 20:", max_of_two(10, 20))
print("Sum of [1, 2, 3, 4, 5]:", sum_list([1, 2, 3, 4, 5]))
```
